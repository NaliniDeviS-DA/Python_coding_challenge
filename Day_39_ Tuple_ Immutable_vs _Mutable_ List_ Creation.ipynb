```python
# Create a tuple with fixed product categories like Electronics, Stationery, and Clothing.
Electronics=('TV','Smart Phone', 'Laptop', 'Projector')
Stationery=('Pen','Pencil','Eraser','Sharpner')
Clothing=('Jean','T-Shirt','Frock','Kurthi')

# Create a list of products that can be updated dynamically.
list_a=list(Electronics)

n=int(input("Enter number of eletronics items: "))

for i in range(n):
  Electronics_list=input(f"Enter Item{i+1}: ")
  list_a.append(Electronics_list)
print("Electronics: ",Electronics)
print("Stationery: ",Stationery)
print("Clothing: ",Clothing)
print("\n Updated list of Electronics:",list_a)

# Ask the user to input the index of a product to remove it from the list.

prod_index=int(input("\n Enter the index of a product to remove it from the list: "))
list_a.pop(prod_index)
print("\n Updated list of Electronics after removal:",list_a)
```
