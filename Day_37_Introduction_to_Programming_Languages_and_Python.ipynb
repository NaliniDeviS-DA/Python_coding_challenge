# Project Title: Student Profile Creator

```python
# studentâ€™s name, age, and their favourite subjects they are enrolled in.

Name=input("Enter name of the student: ")
Age= int(input("Enter age of the student: "))

subjects=[]
n = int(input("How many subjects do u want to enter?: "))

for i in range(n):
  sub = input(f"Enter subject {i+1}: ")
  subjects.append(sub)
print("\n")
print(f"*****Information of Student - {Name.upper()}: *****")
print(f"Age of {Name.upper()} is:{Age}")
print(f"Favourite subjects of {Name} is: {subjects}")
print("\n")
```
