```python
# Ask the user to input a sentence (representing feedback).
# Count how many words are in the sentence.
# Check how many times the word “data” appears in the sentence, regardless of case.
# Extract and print the first 5 and last 5 characters of the feedback separately.
# Reverse the sentence and print it.

FB = input("Enter your feedback plzz!: ")
words=FB.split()
words
print(f"Feedback contains {len(words)} words.")
sentence=FB.lower()
data_count=sentence.split().count("data")
print(f"The word data appears {data_count} times")
first_5_chars=FB[0:5]
last_5_chars=FB[-5:]
print(f"First 5 characters of Feedback is: {first_5_chars}")
print(f"Last 5 characters of Feedback is: {last_5_chars}")
print(f"Reverse of Feedback: {FB[::-1]}")
```
