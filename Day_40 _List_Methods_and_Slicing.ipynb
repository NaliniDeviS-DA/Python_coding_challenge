```python
# Create a 2D list where each row holds a contact’s name, phone number, and email address.
# Allow the user to add a new contact.
# Remove the last item
# Get new items from the users using input() and add those items to the list
# Sort the list in ascending order based on any values

user_list=[]
n=int(input('Enter number of persons: '))

for i in range(1,n+1):
  name=input("Enter ur name: ")
  phone_number=int(input("Enter ur phone_number: "))
  email=input("Enter ur email_id: ")
  user_list.append([name,phone_number,email])
print(user_list)

# Get new items from the users using input() and add those items to the list
a=input("\nStill u want to add more persons(Y/N)? ")
if a=='Y':
  n=int(input('\nEnter number of persons: '))
  for i in range(1,n+1):
    name=input("Enter ur name: ")
    phone_number=input("Enter ur phone_number: ")
    email=input("Enter ur email_id: ")
    user_list.append([name,phone_number,email])
  print(user_list)
  print("Details updated successfully!")

  #Remove last element
  user_list.pop()
  print('After removing last element: ',user_list)

  # Sort the list in ascending order based on any values
  user_list.sort(key=lambda x: x[0])
  print('After sorting in asc order based on names: ',user_list)

```
